```python
import imgaug as ia
import numpy as np
from PIL import Image
import cv2 as cv
from imgaug import augmenters as iaa
import numpy as np
from random import randint, uniform, shuffle
```

```python
def display(img):
    return Image.fromarray(img[:, :, ::-1])

def Add():
    print("add")
    val = randint(-45, 45)
    seq_l.append(iaa.Add(val))

def GaussianBlur():
    print("blur")
    val = uniform(1.0, 3.0)
    seq_l.append(iaa.GaussianBlur(val))

def Hue():
    print("hue")
    val = randint(-10,10)
    seq_l.append(iaa.AddToHueAndSaturation(val))

def Noise():
    print('noise')
    val1 = uniform(0.0, .04)
    seq_l.append(iaa.AdditiveGaussianNoise(scale=val1*225))

def Contrast():
    print('contrast')
    val1 = uniform(.2, .5)
    val2 = uniform(.5, .9)
    seq_l.append(iaa.ContrastNormalization(alpha=(val1, val2)))
    
    
def Crop():
    top = randint(0,10)
    right = randint(0,10)
    bottom = randint(0,10)
    left = randint(0,10)
    print("zoomed by: top=%d, right=%d, bottom=%d, left=%d" %(top, right, bottom, left))
    
def Zoom():
    val = uniform(.8, 1.5)
    print("zoom by: x%f" %val)
    seq_l.append(iaa.Affine(scale=val))
```

```python
seq_l = []
def augment(noise=False, add=False, blur=False, hue=False, contrast=False, crop=False, zoom=False):
    if noise:
        Noise()
    if add:
        Add()
    if blur:
        GaussianBlur()
    if hue:
        Hue()
    if contrast:
        Contrast()
    if crop:
        Crop()
    if zoom:
        Zoom()
    seq = iaa.Sequential(seq_l)

    print("amount of augmentations: " + str(len(seq_l)))

    return seq.augment_image(cv.imread('1.jpg'))
x = augment(noise=1,add=1,blur=1, zoom=1, crop=1)
display(x)
```
